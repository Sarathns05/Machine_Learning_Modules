```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Day=[1,2,3,4,5,6,7]
Temp=[50,51,42,45,48,55,58]
```

```python
plt.plot(Day,Temp)
plt.show()
```

```python
plt.plot(Day,Temp,color='green',linestyle='dashed',marker='o',linewidth=2)
plt.show()
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.plot(Day,Temp,color='green',linestyle='dashdot',marker='o',linewidth=2)
plt.show()
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title("weather info")

plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")
plt.legend(loc='best',shadow=True,fontsize="large")
plt.grid()
plt.show()
```

```python
# Simple bar chart showing revenues of major US tech companies

company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
plt.xlabel("Major companies")
plt.ylabel("Revenue")
plt.title("Revenue model for top companies")

plt.bar(company,revenue,label="Revenue")
plt.bar(company,profit,label="profit")
plt.legend()
plt.show()
```

```python
xpos = np.arange(len(company))
xpos

```

```python
plt.bar(xpos,revenue)
```

```python
plt.xticks(xpos,company)
plt.xlabel("Major companies")
plt.ylabel("Revenue")
plt.title("Revenue model for top companies")

plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="profit")
plt.legend()
plt.show()
```

```python
plt.xticks(xpos,company)
plt.bar(xpos,revenue)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(10)
data = np.random.normal(100,20,200)

data
```

```python
fig = plt.figure(figsize=(10,5))

plt.boxplot(data)
plt.show()
```

```python
np.median(data)
```

```python
np.random.seed(10)
data_1 = np.random.normal(100,10,200)
data_2 = np.random.normal(90,20,200)
data_3 = np.random.normal(80,30,200)
data_4 = np.random.normal(70,40,200)

data = [data_1,data_2,data_3,data_4]
plt.figure(figsize=(10,5))

plt.boxplot(data,vert=False,patch_artist=True)
plt.show()
```

```python
# Draw pie chart to track down home expenses
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]

plt.figure(figsize=(10,5))
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=1.3,autopct="%0.2f%%",shadow=True,explode=[0.1,0,0,0.2,0],startangle=45)
plt.legend()
#plt.show()
plt.savefig("piechart_new.jpg")
```
